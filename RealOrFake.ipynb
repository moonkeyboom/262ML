
### 4. รูปแบบการจัดเก็บ (เช่น format และ โครงสร้าง)
Format: JPG files

CIFAR-10 datasets ข้างในมี 10 classes แต่ละ class มีรูป 6000 รูป ทั้งหมด 60000 รูป 

+ โดยมีข้อมูลแบบ train จำนวน 50000 รูป และ test จำนวน 10000 รูป

+ ข้อมูลแบ่งเป็น 5 training batches และ 1 test batches แต่ละ batch มี 10000 รูป

+ dataset มีทั้งในรูปแบบ Python, Matlab, และ binary โดยข้อมูลเก็บเป็นอาร์เรย์ของ pixel และ labels

โดยข้างใน batch จะเก็บข้อมูลเป็น dictionary โดยมี 2 elements คือ 

1. data
2. labels

ข้อมูลแบ่งออกเป็น 10 labels name คือ 

1. airplane 
2. automobile 
3. bird
