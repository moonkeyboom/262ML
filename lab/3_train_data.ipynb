## Path

```python
from pathlib import Path

# แก้ไขปัญหา path window(\) mac(/)
# Path -> class name
```

```python
Path("cats_vs_dogs")
```

```python
list(Path("cats_vs_dogs/cat").iterdir())
# list ว่าใน path มีอะไรอยู่บ้าง
```

```python
data = {"path" : [], "class": []}  # สร้าง dict เก่บ list
for class_ in Path("cats_vs_dogs").iterdir():
    paths = class_.glob("*") #method ที่ใช้ list ดาม pattern ที่กำหนด "*" regular pattern เอาหมด, "*.jpg" เอา jpg
    for path in paths:
        data["path"].append(path)
        data["class"].append(class_.name) # .name เอาแค่ชื่อ ไม่รวม path

cats_vs_dogs = pd.DataFrame.from_dict(data)
cats_vs_dogs
```
